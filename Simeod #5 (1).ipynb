(a) Let $y(t)$ be the amount of chemical in the tank in lbs at time $t$ in hours.  Let $T(t)$ be the total amount of liquid in the tank in gallons at time $t$ in hours. Then we know that $y'(t) = (.08)(12) - 20\frac{y}{T}$ and $y(0) = 27 lbs$, $T(0) = 1000$, and $T(t) = 1000 + 12t - 20t = 1000 - 8t$.  Thus, $y'(t) = 0.96 - 20(\frac{y}{1000 - 8t})$

(b) Separation of variables won't work here because $y'(t)$ is not separable.  The integrating factors approach works best here because $y'(t)$ is of form $y' = k - \frac{ly}{a + bt}$ which is similar to the product rule.

(c) To solve, take: $$y' + \frac{20}{1000-8t}y = 0.96$$ Then we can set up our integrating factor with: $$d(\mu y) = \mu k$$ So $$\mu y = \int \mu k dt$$  Now we can substitute in $\mu$ to get: $$\frac{d\mu}{dt} = \mu ' = \frac{20}{1000 - 8t}$$ $$\int \mu = \int \frac{20 dt}{1000 - 8t}$$ So $$\mu = -\frac{5}{2}\ln \left|1000-8t\right|+C$$  Thus, $$d(\frac{-\frac{5}{2}\ln \left|1000-8t\right| y}{dt}) = -\frac{5}{2}\ln \left|1000-8t\right| \frac{dy}{dt} -\frac{5}{2}\ln \left|1000-8t\right|y = -2.4\ln \left|1000-8t\right|$$ So, $$y(t) = -2.4\left(t\ln \left|-1000+8t\right|+125-t-125\ln \left|-125+t\right|\right)+C$$

$y(0) = 27$ so $$y(t) = -2.4\left(t\ln \left|-1000+8t\right|+125-t-125\ln \left|-125+t\right|\right)-1121.5$$
